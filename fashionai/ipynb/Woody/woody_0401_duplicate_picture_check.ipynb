```python
import os
import cv2
import random
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from tqdm import tqdm
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
df_train = pd.read_csv('label.csv', header=None)
df_train.columns = ['image_id', 'class', 'label']
df_train.head()
```

```python
classes = ['collar_design_labels', 'neckline_design_labels', 'skirt_length_labels', 
           'sleeve_length_labels', 'neck_design_labels', 'coat_length_labels', 'lapel_design_labels', 
           'pant_length_labels']
len(classes)
```

思路：  
1、对每个class进行遍历；  
2、找出每个class的重复图片；   
3、判断重复图片的对应的标签是否一致，如果不一致的打印出来;  

```python
#加载所有的class
dict_class_dupilcate={}#将所有重复图片的的索引记录下来
dict_class_diff={}#将所有重复图片但是y值不一样的索引记录下来
for i in range(len(classes)):
    print('this is %s class'%i)
    cur_class = classes[i]#对每个class分别计算
    df_load = df_train[(df_train['class'] == cur_class)].copy()
    df_load.reset_index(inplace=True)
    del df_load['index']
    #print('{0}: {1}'.format(cur_class, len(df_load)))
    n = len(df_load) # 样本数量
    n_class = len(df_load['label'][0]) #label的值
    width = 50

    X = np.zeros((n, width, width, 3), dtype=np.uint8)
    y = np.zeros((n, n_class), dtype=np.uint8)
   #加载图片
    for j in range(n): 
        tmp_label = df_load['label'][j]
        if len(tmp_label) > n_class:
            print(df_load['image_id'][j])
        #加载图片数据,且将图片的大小从（512，512，3）变为（50，50，3），这样运算起来比较节省时间
        X[j] = cv2.resize(cv2.imread('base/{0}'.format(df_load['image_id'][j])), (width, width))
        y[j][tmp_label.find('y')] = 1#加载label
        
    #找出重复的图片    
    num_duplicate=[]
    for n in tqdm(range(len(X))):
        for m in range(len(X)):
            if n !=m:
                if ((X[n] ==X[m]).all())==True:
                    num_duplicate.append((n,m))
    #找出重复图片且标记值不一样的               
    num_diff=[]
    for p in num_duplicate:
        if (y[p[0]] ==y[p[1]]).all()==False:
            num_diff.append(p[0])
            
    dict_class_dupilcate[i]=num_duplicate
    dict_class_diff[i]=num_diff
    
print('dict_class_dupilcate is :') 
print(dict_class_dupilcate)

print('dict_class_diff is :') 
print(dict_class_diff)
```
