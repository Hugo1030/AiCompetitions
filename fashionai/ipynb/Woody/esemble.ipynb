```python
def avg_result(label1, label2):
    lst1 = []
    for i in label1['result'].values:
        lst = i.split(';')
        lst1.append(lst)
        
    lst2 = []
    for i in label2['result'].values:
        lst = i.split(';')
        lst2.append(lst)
        
    m = len(lst1)
    n = len(lst1[0])
    
    result = []

    for i in range(m):
        tmp_result = ''
        for j in range(n):
            avg = (float(lst1[i][j])*0.5 + float(lst2[i][j])*0.5)
            tmp_result += '{:.4f};'.format(avg)
        result.append(tmp_result[:-1])
    return result
```

```python
label1= pd.read_csv(
    '../../raw/Woody/result/neck_0602_inception_psd_all.csv',header=None)
label1.columns = ['image_id', 'class', 'result']
```

```python
label2=pd.read_csv('../../raw/Woody/result/neck_0602_xception_psd_all.csv',header=None)
label2.columns = ['image_id', 'class', 'result']
```

```python
result=avg_result(label1, label2)
```
