```python
import pandas as pd
import numpy as np
```

```python
df_pant1 = pd.read_csv('./result/{}_0418a.csv'.format('pant'), header=None) # xception切分训练集
df_pant2 = pd.read_csv('./result/{}_0418alldata.csv'.format('pant'), header=None) # InceptionResNetV2全训练集

df_lapel1 = pd.read_csv('./result/{}_0418a.csv'.format('lapel'), header=None) # xception切分训练集
df_lapel2 = pd.read_csv('./result/{}_0418alldata.csv'.format('lapel'), header=None) # InceptionResNetV2全训练集
```

```python
# 给分类别增添表头
df_pant1.columns = ['image_id', 'class', 'result']
df_pant2.columns = ['image_id', 'class', 'result']

df_lapel1.columns = ['image_id', 'class', 'result']
df_lapel2.columns = ['image_id', 'class', 'result']
```

```python
df_lapel2
```

```python
def avg_result(label1, label2):
    lst1 = []
    for i in label1['result'].values:
        lst = i.split(';')
        lst1.append(lst)
        
    lst2 = []
    for i in label2['result'].values:
        lst = i.split(';')
        lst2.append(lst)
        
    m = len(lst1)
    n = len(lst1[0])
    
    result = []

    for i in range(m):
        tmp_result = ''
        for j in range(n):
            avg = ((float(lst1[i][j]) + float(lst2[i][j]))/2)
            tmp_result += '{:.4f};'.format(avg)
        result.append(tmp_result[:-1])
    return result
```

```python
pant_avg = avg_result(df_pant1, df_pant2)
lapel_avg = avg_result(df_lapel1, df_lapel2)
```

```python
df_pant1['result'] = pant_avg
df_lapel1['result'] = lapel_avg
```

```python
df_pant1.to_csv('./result/{}_0418b.csv'.format('pant'), header=None, index=False)
df_lapel1.to_csv('./result/{}_0418b.csv'.format('lapel'), header=None, index=False)
```
