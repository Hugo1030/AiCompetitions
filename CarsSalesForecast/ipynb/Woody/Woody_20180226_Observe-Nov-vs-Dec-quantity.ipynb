```python
for i in class_id:
#     list2013=[]
#     list2014=[]
#     list2015=[]
#     list2016=[]
    if float(data[(data.class_id==i)&(data.sale_date=='2013-10')]['sale_quantity']) !=0.0:#10月份有产量，11月份和12月份肯定就有
        pd_2013_10=data[(data.class_id==i)&(data.sale_date=='2013-10')]['sale_quantity']
        pd_2013_11=data[(data.class_id==i)&(data.sale_date=='2013-11')]['sale_quantity']
        pd_2013_12=data[(data.class_id==i)&(data.sale_date=='2013-12')]['sale_quantity']
        data_2013[i]=[float(pd_2013_10),float(pd_2013_11),float(pd_2013_12)]
        
    if float(data[(data.class_id==i)&(data.sale_date=='2014-10')]['sale_quantity']) !=0.0:
        pd_2014_10=data[(data.class_id==i)&(data.sale_date=='2014-10')]['sale_quantity']
        pd_2014_11=data[(data.class_id==i)&(data.sale_date=='2014-11')]['sale_quantity']
        pd_2014_12=data[(data.class_id==i)&(data.sale_date=='2014-12')]['sale_quantity']
        data_2014[i]=[float(pd_2014_10),float(pd_2014_11),float(pd_2014_12)]
        
    if float(data[(data.class_id==i)&(data.sale_date=='2015-10')]['sale_quantity']) !=0.0:
        pd_2015_10=data[(data.class_id==i)&(data.sale_date=='2015-10')]['sale_quantity']
        pd_2015_11=data[(data.class_id==i)&(data.sale_date=='2015-11')]['sale_quantity']
        pd_2015_12=data[(data.class_id==i)&(data.sale_date=='2015-12')]['sale_quantity']
        data_2015[i]=[float(pd_2015_10),float(pd_2015_11),float(pd_2015_12)]
        
    if float(data[(data.class_id==i)&(data.sale_date=='2016-10')]['sale_quantity']) !=0.0:
        pd_2016_10=data[(data.class_id==i)&(data.sale_date=='2016-10')]['sale_quantity']
        pd_2016_11=data[(data.class_id==i)&(data.sale_date=='2016-11')]['sale_quantity']
        pd_2016_12=data[(data.class_id==i)&(data.sale_date=='2016-12')]['sale_quantity']
        data_2016[i]=[float(pd_2016_10),float(pd_2016_11),float(pd_2016_12)]
        
        
    
```
